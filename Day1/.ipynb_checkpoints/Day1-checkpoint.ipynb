```python
import pandas as pd
import numpy as np
import itertools
```

```python
# Import data
#data = pd.read_csv('input.txt', header=None)
with open('input.txt', 'r') as f:
    data =  [int(line.strip()) for line in f]
```

```python
#Part one
comb = list(itertools.product(data,data))
for item in comb:
    if sum(list(item)) == 2020:
        print(item)
        print(np.prod(item))
```

```python
#Part two
comb = list(itertools.product(data,data,data))
for item in comb:
    if sum(list(item)) == 2020:
        print(item)
        print(np.prod(item))
```
