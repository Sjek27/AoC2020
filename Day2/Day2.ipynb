```python
import pandas as pd
import numpy as np
import itertools
```

```python
# Load data
data = pd.read_csv('input.txt', sep=" ", header=None)
data.columns = ['count', 'letter', 'word']
data['letter'] = data['letter'].str.replace(':','')
```

# Part1

```python
data['count_letter'] = data.apply(lambda x: x['word'].count(x['letter']), axis=1)
```

```python
data[['min','max']] = data['count'].str.split('-', expand=True)
```

```python
data
```

```python
sum(data.apply(lambda x: x['count_letter'] in range(int(x['min']), int(x['max'])+1), axis=1))
```

# Part2

```python
data['pos1'] = data.apply(lambda x: x['word'][int(x['min'])-1] == x['letter'], axis=1)
data['pos2'] = data.apply(lambda x: x['word'][int(x['max'])-1] == x['letter'], axis=1)
```

```python
data['valid'] = data['pos1'] ^ data['pos2'] # ^ is xor for two booleans
```
